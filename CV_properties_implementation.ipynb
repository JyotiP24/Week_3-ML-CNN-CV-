```python

def get_limits(color):
    c = np.uint8([[color]])  # BGR values
    hsvC = cv2.cvtColor(c, cv2.COLOR_BGR2HSV)
    print(f"HSV value of the target color: {hsvC}")  # Print HSV value for debugging

    hue = hsvC[0][0][0]  # Get the hue value

    # Handle green hue range manually
    lower_limit = np.array([35, 100, 100], dtype=np.uint8)
    upper_limit = np.array([85, 255, 255], dtype=np.uint8)

    return lower_limit, upper_limit

# Define the target color (BGR format)
target_color = [0, 255, 0]  # For  green

# Get the HSV color limits for the target color
lower_limit, upper_limit = get_limits(target_color)

# Convert the image from BGR to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Create a mask using the color limits
mask = cv2.inRange(hsv_img, lower_limit, upper_limit)

# Apply the mask to the original image
cartoon_y_mentee2= cv2.bitwise_and(img, img, mask=mask)
cartoon_y_mentee=(cv2.cvtColor(cartoon_y_mentee, cv2.COLOR_BGR2RGB))


```
